In this notebook we look at BigPatent Dataset. The dataset is divided into 9 Cooperative Patent Classification (CPC) codes. Each code category dataset has three splits as train, validation and test. Each split has two columns "description" and "abstract". We calculate length of "description" column to find category which has lowest average length.

The category with lowest average length for "description" column will be a good candidate for use in TLDR project.

Since dataset cannot be loaded all at once in the colab, we calculate length separately for all categories.

```python
!pip install datasets
from datasets import load_dataset
import pandas as pd
```

```python
dict_ds = {}
dict_train = {}
dict_ds['a'] = load_dataset("big_patent", "a")  # only 'a' CPC codes
dict_train['a'] = pd.DataFrame(dict_ds['a']['train'])
dict_train['a']['len_char']=dict_train['a']['description'].apply(lambda x: len(x))
dict_train['a']['len_words']=dict_train['a']['description'].apply(lambda x: len(x.split()))
print("characters = ",sum(dict_train['a']['len_char'])/len(dict_train['a']['len_char']))
print("words = ",sum(dict_train['a']['len_words'])/len(dict_train['a']['len_words']))
#characters = 31934.67609427223
#words = 4795.441837894955
```
